### Tokenization

```python
# tokenize to word
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
word_tokens = word_tokenize(speech_text.lower())
print("words length : ", len(word_tokens))
```

### Bigram, Trigram, Ngram

```python
speech_bigrams = list(nltk.bigrams(word_tokens))
speech_trigrams = list(nltk.trigrams(word_tokens))
speech_ngrams = list(nltk.ngrams(word_tokens, 5))
```

### Most common word

```python
# How many words are repeated ?
from nltk.probability import FreqDist
freq_words = FreqDist(word_tokens)
print(freq_words)

# word frequency distribution
freq_words.most_common(10)
```

### Stopwords - Remove Stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words=set(stopwords.words("english"))
filtered_word = word_tokens[:]

for word in word_tokens:
  if word in stop_words:
    # remove every stop words in filtered words
    filtered_word.remove(word)
print(filtered_word)
```

```python
# remove punctuations from filtered word
import re
punctuation = re.compile(r'[-.?!,:;()|0-9]')
post_punctuation = []
for words in filtered_word:
  word = punctuation.sub("", words)
  if(len(word) > 0):
    post_punctuation.append(word)
print(post_punctuation)
```

### Stemming

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
snowball_stemmer = SnowballStemmer('english')
root_words = []

for word in post_punctuation:
  # Choose stemmer algorithm
  word_stem = porter_stemmer.stem(word)
  root_words.append(word_stem)
print(root_words)
```

### Lemmatization

```python
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
lemm_words = []

for word in post_punctuation:
  lemm_words.append(lemmatizer.lemmatize(word, pos='v'))
print(lemm_words)
```

### Most Common Word

```python
# How many words are repeated ?
freq_words_final = FreqDist(lemm_words)
print(freq_words_final)

# word frequency distribution
freq_words_final.most_common(10)
```

### POS Tags

```python
nltk.download('averaged_perceptron_tagger')

for token in lemm_words:
    print(nltk.pos_tag([token]))
```
